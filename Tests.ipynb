```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
omega = np.linspace(-10, 10, 1000)
T1 = 2
T2 = 1
Ts = np.sqrt(T1 * T2)
Rse = 10
Kxy = omega / Ts / (1 + (omega / Ts) ** 2) * Rse / Ts
```

```python
plt.figure(figsize=(12, 6))
plt.plot(omega, Kxy)
plt.scatter(Ts, Kxy[np.argmin(np.abs(omega - Ts))], color='red')
plt.xlabel(r'$\Omega_d $')
plt.ylabel(r'$ \vert K_xy \vert $')
plt.show()
```
